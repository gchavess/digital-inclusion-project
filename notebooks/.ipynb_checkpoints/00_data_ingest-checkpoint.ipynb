```python
# 00_data_ingest.ipynb
import pandas as pd
from pathlib import Path


DATA_PATH = Path('..') / 'data' / 'internet_adoption.csv'


df = pd.read_csv(DATA_PATH)


# Visão inicial
print(df.shape)
df.head()


# Tipos e nulos
print(df.info())
print(df.isna().sum())


# Exemplos de limpeza
# - converter colunas numéricas
num_cols = ['internet_users_percent', 'gdp_per_capita', 'population']
for c in num_cols:
df[c] = pd.to_numeric(df[c], errors='coerce')


# - preencher ou remover NA (exemplo: remover linhas com target NA)
df = df.dropna(subset=['internet_users_percent'])


# - salvar versão tratada
df.to_csv('../data/internet_adoption_clean.csv', index=False)
```
