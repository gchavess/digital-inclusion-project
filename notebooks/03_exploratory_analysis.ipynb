```python
# ðŸ”¹ 1. Imports e carregamento

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")

df = pd.read_csv("processed_data.csv")

print("Dataset carregado:", df.shape)
df.head()
```

```python
# ðŸ”¹ 2. EstatÃ­sticas descritivas
df.describe(include='all').T
```

```python
# ðŸ”¹ 3. DistribuiÃ§Ã£o das principais variÃ¡veis

cols_hist = [
    "Internet_Penetration (%)",
    "Total_Speed_Index",
    "Digital_Inclusion_Index",
    "Relative_Internet_Cost",
    "Cloud_Service_Adoption (%)",
]

df[cols_hist].hist(bins=25, figsize=(14, 10))
plt.suptitle("DistribuiÃ§Ã£o das VariÃ¡veis Principais", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# ðŸ”¹ 4. Heatmap de CorrelaÃ§Ã£o

# Seleciona somente colunas numÃ©ricas
df_num = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(18, 14))
sns.heatmap(df_num.corr(), cmap="coolwarm", annot=False)
plt.title("Matriz de CorrelaÃ§Ã£o")
plt.show()
```

```python
# ðŸ”¹ 5. Top 10 correlaÃ§Ãµes com PenetraÃ§Ã£o de Internet

# Filtra somente colunas numÃ©ricas
df_num = df.select_dtypes(include=['float64', 'int64'])

# Calcula correlaÃ§Ã£o apenas entre variÃ¡veis numÃ©ricas
corr_target = df_num.corr()["Internet_Penetration (%)"].sort_values(ascending=False)

# Mostra top 10
corr_target.head(10).to_frame("Correlation").style.background_gradient(cmap="coolwarm")
```

```python
# ðŸ”¹ 6. Pairplot das variÃ¡veis mais importantes

sns.pairplot(
    df[
        [
            "Internet_Penetration (%)",
            "Digital_Inclusion_Index",
            "Total_Speed_Index",
            "GDP_Per_Capita (USD)",
            "Cloud_Service_Adoption (%)",
        ]
    ],
    diag_kind="kde"
)
plt.show()
```

```python
# ðŸ”¹ 7. Boxplots por regiÃµes urbanas/rurais

plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df,
    x="Urban_Population_Split (%)",
    y="Internet_Penetration (%)",
    hue="Digital_Literacy_Score (%)",
    palette="viridis"
)
plt.title("PenetraÃ§Ã£o da Internet vs PopulaÃ§Ã£o Urbana")
plt.show()
```

```python
# ðŸ”¹ 8. Scatterplot: renda vs penetraÃ§Ã£o

plt.figure(figsize=(10, 6))
sns.regplot(
    data=df,
    x="GDP_Per_Capita (USD)",
    y="Internet_Penetration (%)",
    scatter_kws={"alpha": 0.5}
)
plt.title("RelaÃ§Ã£o entre PIB per Capita e Internet Penetration")
plt.show()
```

```python
# ðŸ”¹ 9. AnÃ¡lise de Custo Relativo da Internet

plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df,
    x="Relative_Internet_Cost",
    y="Internet_Penetration (%)",
    hue="GDP_Per_Capita (USD)",
    palette="coolwarm"
)
plt.title("Custo Relativo da Internet vs PenetraÃ§Ã£o")
plt.show()
```

```python
# ðŸ”¹ 10. AnÃ¡lise de distribuiÃ§Ã£o geogrÃ¡fica

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x="Longitude",
    y="Latitude",
    size="Internet_Penetration (%)",
    hue="Total_Speed_Index",
    sizes=(20, 400),
    palette="viridis"
)
plt.title("DistribuiÃ§Ã£o GeogrÃ¡fica: PenetraÃ§Ã£o e Velocidade")
plt.show()
```

```python
# ðŸ”¹ 11. AnÃ¡lise de polÃ­ticas governamentais

plt.figure(figsize=(10,6))
sns.regplot(
    data=df,
    x="Government_Digital_Policy_Index (%)",
    y="Internet_Penetration (%)",
    scatter_kws={"alpha":0.4}
)
plt.title("Efeito das PolÃ­ticas Governamentais na AdoÃ§Ã£o de Internet")
plt.show()
```
